## **Matplotlib Plotting**
**plot():** function used to draw points(markers) in a diagram.<br>
This function takes parameters for specifying points in the diagram.<br>
* Parameter 1 is an array containing the points on the x-axis.
* Parameter 2 is an array containing the points on the y-axis.

```python
#draw line from (1, 3) to (6, 10)
import matplotlib.pyplot as plt
import numpy as np

xpts = ([1, 6])
ypts = ([3, 10])

plt.plot(xpts, ypts)
plt.show()
```

```python
#plotting without line
import matplotlib.pyplot as plt
import numpy as np

xpts = np.array([1, 8])
ypts = np.array([3, 10])

plt.plot(xpts, ypts, 'o')
plt.show()
```

```python
#multiple points from (1, 3) to (2, 8), (6, 1) and (8, 10)
import matplotlib.pyplot as plt
import numpy as np

xpts = np.array([1, 2, 6, 8])
ypts = np.array([3, 8, 1, 10])

plt.plot(xpts, ypts)
plt.show()
```

**Default x-points**<br>
If points on the x-axis are not specified, they will get the default values 0, 1, 2 etc., depending on the length of the y-points

```python
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypts)
plt.show()
```

### **Matplotlib Markers**

```python
import matplotlib.pyplot as plt
import numpy as np

ypts = ([5, 8, 11, 1, 7])

plt.plot(ypts, marker = 'o')
plt.show
```

**Format strings fmt**<br>
You can also use the shortcut string notation parameter to specify the marker. In the following syntax:<br>
***marker|line|color***

```python
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([2, 8, 5, 3])

plt.plot(ypts, 'o:r')
plt.show
```

**Marker Size**<br>
Argument *markersize* or *ms* to set the size of the markers.

```python
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([4, 2, 7, 1])

plt.plot(ypts, marker = '*', ms = 20)
plt.show
```

**Marker Color**<br>
Argument *markeredgecolor* or *mec* to set the color of the edge of the markers

```python
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([2, 6, 4, 9])

plt.plot(ypts, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

**Marker Face Color**<br>
Argument *markerfacecolor* or *mfc* to set the color inside the edge of the markers.

```python
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([4, 2, 9, 5])

plt.plot(ypts, marker='o', ms=20, mfc='r')
plt.show()
```

```python
#set the color of face and edge to one color
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([8, 3, 5, 2])

plt.plot(ypts, marker='o', mec='m', mfc='m', ms=15)
plt.show()
```

```python
#One can use hexadecimal values
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([8, 3, 5, 2])

plt.plot(ypts, marker='o', mec='#c3c3c3', mfc='#c3c3c3', ms=12)
plt.show()
```

## **Matplotlib Line**
**Linestyle**<br>
Argument *Linestyle* or *ls* to change he style of the plotted line.<br>
Shorter syntax<br>
* *linestyle* can be written as *ls*
* *solid* can be written as *-*
* *dotted* can be written as *:*
* *dashed* can be written as *--*
* *dashdot* can be written as *-.*
* *none* can be written as *''* or *' '*

```python
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([3, 8, 1, 10])

plt.plot(ypts, ls='dotted')
plt.show()
```

**Line Color**<br>
Argument *color* or *c* to set the color of the line.

```python
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([3, 8, 1, 10])

plt.plot(ypts, c='g')
plt.show()
```

```python
#you can use hexadecimals values
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([3, 8, 1, 10])

plt.plot(ypts, c='#FFD700')
plt.show()
```

**Line Width**<br>
Argument *linewidth* or *lw*, to change the width of the line.

```python
import matplotlib.pyplot as plt 
import numpy as np

ypts = np.array([3, 8, 1, 10])

plt.plot(ypts, lw='12')
plt.show()
```

**Multiple Lines**<br>
Plot as many lines as possible simply by adding more *plt.plot()* functions

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 2, 7])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

You can also plot many lines by adding the points for the x- and y-axis for each line in the same plt.plot() function.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
