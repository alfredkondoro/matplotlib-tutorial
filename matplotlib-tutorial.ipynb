## **Matplotlib Plotting**
**plot():** function used to draw points(markers) in a diagram.<br>
This function takes parameters for specifying points in the diagram.<br>
* Parameter 1 is an array containing the points on the x-axis.
* Parameter 2 is an array containing the points on the y-axis.

```python
#draw line from (1, 3) to (6, 10)
import matplotlib.pyplot as plt
import numpy as np

xpts = ([1, 6])
ypts = ([3, 10])

plt.plot(xpts, ypts)
plt.show()
```

```python
#plotting without line
import matplotlib.pyplot as plt
import numpy as np

xpts = np.array([1, 8])
ypts = np.array([3, 10])

plt.plot(xpts, ypts, 'o')
plt.show()
```

```python
#multiple points from (1, 3) to (2, 8), (6, 1) and (8, 10)
import matplotlib.pyplot as plt
import numpy as np

xpts = np.array([1, 2, 6, 8])
ypts = np.array([3, 8, 1, 10])

plt.plot(xpts, ypts)
plt.show()
```

**Default x-points**<br>
If points on the x-axis are not specified, they will get the default values 0, 1, 2 etc., depending on the length of the y-points

```python
import matplotlib.pyplot as plt
import numpy as np

ypts = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypts)
plt.show()
```
